## boxplot and violinplot

boxplots og violinplots brukes til å vise fordelingen av kategoriske data. En boksplott (eller boks-og-whisker-plott) viser fordelingen av kvantitative data på en måte som gjør det mulig å sammenligne mellom variabler eller på tvers av nivåer av en kategorisk variabel. Ruten viser kvartilene i datasettet mens kinnskjeggene strekker seg for å vise resten av distribusjonen, bortsett fra punkter som er bestemt for å være "outliers" ved hjelp av en metode som er en funksjon av interkvartilområdet.

### violinplot

Et fiolinplott spiller en lignende rolle som en boks og kinneplott. Den viser fordelingen av kvantitative data over flere nivåer av en (eller flere) kategoriske variabler slik at disse fordelingen kan sammenlignes. I motsetning til en boksplott, der alle plottkomponentene tilsvarer faktiske datapunkter, har fiolinplottet en estimering av kjernetetthet av den underliggende fordelingen.

## stripplot and swarmplot

Stripplottet tegner et scatterplot der en variabel er kategorisk. Et stripeplott kan tegnes alene, men det er også et godt supplement til en boks eller fiolinplott i tilfeller der du vil vise alle observasjoner sammen med en viss fremstilling av den underliggende distribusjonen.

Svermplottet ligner på stripplot (), men punktene justeres (bare langs den kategoriske aksen) slik at de ikke overlapper hverandre. Dette gir en bedre representasjon av fordelingen av verdier, selv om den ikke skaleres like godt til et stort antall observasjoner (både når det gjelder muligheten til å vise alle punktene og når det gjelder beregningen som trengs for å ordne dem).
