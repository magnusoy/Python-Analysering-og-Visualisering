## lmplot()

### Arbeide med markører

lmplot kwargs blir sendt videre til **regplot** som er en mer generell form for lmplot(). regplot har en scatter_kws-parameter som blir sendt til plt.scatter. Så du vil sette parameteren s i den ordboken, som tilsvarer (litt forvirrende) den kvadratiske markørstørrelsen. Med andre ord ender du opp med å sende en ordbok med de grunnleggende matplotlib-argumentene, i dette tilfellet s for størrelsen på et spredningsdiagram. Generelt vil du sannsynligvis ikke huske dette fra toppen av hodet, men i stedet referere til dokumentasjonen.
